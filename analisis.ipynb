## Inspección de Datos - Formula 1

El objetivo es inspeccionar la estructura, calidad de los datos y realizar una limpieza inicial para preparar el modelo en estrella.

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# Lista de archivos CSV disponibles
files = [f for f in os.listdir('.') if f.endswith('.csv')]
print(f"Archivos encontrados: {len(files)}")
files
```

## 1. Inspección General de Archivos
Vamos a cargar cada archivo para ver su cantidad de registros, columnas y los primeros datos.

```python
summary_data = []

for file in files:
    # Nota: Usamos na_values='\\N' porque Ergast usa esa cadena para representar nulos
    df_temp = pd.read_csv(file, na_values='\\N')
    summary_data.append({
        'Archivo': file,
        'Filas': df_temp.shape[0],
        'Columnas': df_temp.shape[1],
        'Nombres Col': list(df_temp.columns)
    })

df_summary = pd.DataFrame(summary_data)
df_summary
```
